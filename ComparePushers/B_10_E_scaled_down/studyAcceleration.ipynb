```python
def plot_hist(track, t=-1, mask = False, fig=None, ax=None):
    if fig is None or ax is None:
        fig, ax = plt.subplots()
    
    label = ("$t = {:.2f}$".format(track["t"][0, t]) + "$[{}]$".format(track.units["t"]))

    data = track["p3"][:,t]
    if mask:
        mask = (data < 1) & (data > -1)
        data = data[mask]

    ax.hist(data, bins=50, label=label)
    ax.set_xlabel("${}$".format(track.labels["p3"]) + "$[{}]$".format(track.units["p3"]))

    plt.legend()

    return fig, ax
```

```python
fig, ax = None, None
fig, ax = plot_hist(sim_Boris_1["test_electrons"]["tracks"], t=-1, fig=fig, ax=ax)
ax.set_title(r"Boris, $2\pi dt w\_{ce} = 1$")
plt.show()
```
