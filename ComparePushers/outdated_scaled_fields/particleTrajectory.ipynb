```python
def plot_trajectory(traj):
    '''
    Function to plot a 3D trajectory over time.
    
    Input:
        - traj: A structured numpy array of shape (num_times,) with named fields "x1", "x2", and "x3".
            numpy.ndarray with dtype containing fields "x1", "x2", "x3"
    
    Output:
        - A 3D plot showing the trajectory.
    
    Example:
        >>> traj = track._data[0, :][["x1", "x2", "x3"]]
        >>> plot_trajectory(traj)
    '''
    
    x, y, z = traj["x1"], traj["x2"], traj["x3"]
    
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot the trajectory
    ax.plot(x, y, z, marker='o', linestyle='-', color='black', markersize=4, label='Trajectory')
    
    # Highlight start and end points
    ax.scatter(x[0], y[0], z[0], color='g', s=50, label='Start')
    ax.scatter(x[-1], y[-1], z[-1], color='r', s=50, label='End')
    
    # Labels and title
    ax.set_xlabel(r'$x_1$')
    ax.set_ylabel(r'$x_2$')
    ax.set_zlabel(r'$x_3$')
    ax.legend()

    return fig, ax
```

```python
def compare_trajectories_3d(trajectories, labels, fig=None, ax=None):
    if fig is None:
        fig = plt.figure(figsize=(10, 7))
    if ax is None:
        ax = fig.add_subplot(111, projection='3d')

    label1 = "start"
    label2 = "end"

    for i, traj in enumerate(trajectories):
        x, y, z = traj["x1"], traj["x2"], traj["x3"]

        if i:
            label1 = None
            label2 = None
        if labels[i] == "Baseline":
            color = "black"
            markersize = 0
        else:
            color = None
            markersize = 1.5

        # Highlight start and end points
        ax.scatter(x[0], y[0], z[0], color='lightgreen', s=20, label=label1)
        ax.scatter(x[-1], y[-1], z[-1], color='r', s=20, label=label2)

        # Plot the trajectory
        ax.plot(x, y, z, marker='o', linestyle='-', markersize=markersize, linewidth=0.8, label=labels[i], color=color)
        


    # Labels and title
    ax.set_xlabel(r'$x_1$')
    ax.set_ylabel(r'$x_2$')
    ax.set_zlabel(r'$x_3$')
    ax.legend()

    return fig, ax
```

```python
# Boris pusher trajectories
particle = 700

traj = [ track_0_1.data[particle, ::10][["x1", "x2", "x3"]], track_1.data[particle, ::1][["x1", "x2", "x3"]], track_10.data[particle, :][["x1", "x2", "x3"]]]
labels = [ r"$2\pi dt w_{ce} = 0.1$", r"$2\pi dt w_{ce} = 1$", r"$2\pi dt w_{ce} = 10$"]



fig, ax = compare_trajectories_3d(traj, labels)

ax.set_title("Boris pusher, particle " + str(particle))
plt.tight_layout()
plt.show()


fig, ax = plt.subplots()
ax.set_ylabel("${}$".format(track_0_1.labels["p3"]) + "$[{}]$".format(track_0_1.units["p3"]))
ax.set_xlabel("${}$".format(track_0_1.labels["t"]) + "$[{}]$".format(track_0_1.units["t"]))
ax.plot(track_0_1.data[particle, ::10][["t"]], track_0_1.data[particle, ::10][["p3"]])
ax.set_title("Particle " + str(particle))
plt.tight_layout()
plt.show()
```
