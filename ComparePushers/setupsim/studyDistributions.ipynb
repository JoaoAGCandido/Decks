```python
def plot_hist(track, t=-1, fig=None, ax=None):
    if fig is None or ax is None:
        fig, ax = plt.subplots()
    
    label = ("$t = {:.2f}$".format(track["t"][0, t]) + "$[{}]$".format(track.units["t"]))
    ax.hist(track["p3"][:,t], bins=50, label=label)
    ax.set_xlabel("${}$".format(track.labels["p3"]) + "$[{}]$".format(track.units["p3"]))

    plt.legend()

    return fig, ax
```

```python
fig, ax = plot_hist(sim_original_fix_bg["test_electrons"]["tracks"], t=1)
fig, ax = plot_hist(sim_original_fix_bg["test_electrons"]["tracks"], t=0, fig = fig, ax = ax)
ax.set_title("no scale, dt = 0.285")
plt.show()
```
