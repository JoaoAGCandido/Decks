```python
import osiris_utils as ou
import numpy as np
from pathlib import Path
from tqdm import tqdm
import matplotlib.pyplot as plt

```

```python
sim_E_0_001 = ou.Simulation("/home/exxxx5/Tese/Decks/ComparePushers/Study_E_scale/E_0_001/Boris_0_1.in")
sim_E_0_01 = ou.Simulation("/home/exxxx5/Tese/Decks/ComparePushers/Study_E_scale/E_0_01/Boris_0_1.in")
sim_E_0_1 = ou.Simulation("/home/exxxx5/Tese/Decks/ComparePushers/Study_E_scale/E_0_1/Boris_0_1.in")
sim_E_1 = ou.Simulation("/home/exxxx5/Tese/Decks/ComparePushers/Study_E_scale/E_1/Boris_0_1.in")
sim_E_10 = ou.Simulation("/home/exxxx5/Tese/Decks/ComparePushers/Study_E_scale/E_10/Boris_0_1.in")
```

```python
def plot_Ene(sim, label, fig=None, ax=None, unload=False):
    step = 100
    
    sim["test_electrons"]["tracks"].load_all()
    track = sim["test_electrons"]["tracks"]
    t = track["t"][0, ::step]

    if fig is None or ax is None:
        fig, ax = plt.subplots()

    Ene = np.average(track["ene"][:, ::step], axis = 0)
    ax.plot(t, Ene, label=label)
    ax.set_xlabel("${}$".format(track.labels["t"]) + "$[{}]$".format(track.units["t"]))
    ax.set_ylabel("${}$".format(track.labels["ene"]) + "$[{}]$".format(track.units["ene"]))

    if unload:
        track.unload()

    return fig, ax
```

```python
fig, ax = None, None
fig, ax = plot_Ene(sim_E_0_001, r"$E_{scale} = 0.001$", fig=fig, ax=ax, unload=True)
fig, ax = plot_Ene(sim_E_0_01, r"$E_{scale} = 0.01$", fig=fig, ax=ax, unload=True)
fig, ax = plot_Ene(sim_E_0_1, r"$E_{scale} = 0.1$", fig=fig, ax=ax, unload=True)
```

```python
fig, ax = plot_Ene(sim_E_1, r"$E_{scale} = 1$", fig=fig, ax=ax, unload=True)
fig, ax = plot_Ene(sim_E_10, r"$E_{scale} = 10$", fig=fig, ax=ax, unload=True)

ax.legend()
ax.set_yscale("log")
ax.set_title("Average Energy")
```
